Given a folder containing .csv files and an output folder, creates a subfolder for each file (named after the file) and moves the file into it. Skips files whose target folder already exists.

```python
from pathlib import Path
import shutil
```

```python
input_folder = Path(r'/media/alberto.vicente/DATA/Ultrasonidos')
output_folder = Path(r'/home/alberto.vicente/phd/DataDriven_UT_AlbertoVicente/03_UT_data/Fabricacion Nacho/medidas/5MHz')
```

```python
vsc_files = list(input_folder.glob('*.vsc'))
print(f'Found {len(vsc_files)} .vsc files')
```

```python
for csv_file in vsc_files:
    target_folder = output_folder / csv_file.stem

    if target_folder.exists():
        print(f'Skipping {csv_file.name} — folder already exists')
        continue

    target_folder.mkdir(parents=True)
    shutil.move(str(csv_file), str(target_folder / csv_file.name))
    print(f'Moved {csv_file.name} → {target_folder}')
    print('')
```
