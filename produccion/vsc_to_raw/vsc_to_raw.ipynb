# VSC to RAW

Recursively searches a folder for `.vsc` files and converts each one to a flat `.raw` volume plus a companion `_ascan_info.txt` file. Files whose `.raw` already exists are skipped.

Output files are written next to the source `.vsc` file using the naming convention `<stem>_ascan{X}x{Y}x{Z}.raw`.

```python
from pathlib import Path
import sys

sys.path.append(r'\\192.168.10.106\imdea\DataDriven_UT_AlbertoVicente\10_code\UTvsXCT-preprocessing')
sys.path.append(r'/home/alberto.vicente/phd/DataDriven_UT_AlbertoVicente/10_code/UTvsXCT-preprocessing')

from preprocess_tools.vsc_to_raw import vsc_to_raw
```

## 1. User inputs

```python
INPUT_DIR = Path(r'/home/alberto.vicente/phd/DataDriven_UT_AlbertoVicente/03_UT_data/Fabricacion Nacho/medidas/5MHz')
```

## 2. Convert

```python
vsc_files = sorted(INPUT_DIR.rglob('*.vsc'))
print(f'Found {len(vsc_files)} .vsc file(s)')

converted, skipped = 0, 0

for vsc_path in vsc_files:
    existing = list(vsc_path.parent.glob(f'{vsc_path.stem}_ascan*.raw'))
    if existing:
        print(f'Skip  {vsc_path.name}  →  {existing[0].name} already exists')
        skipped += 1
        continue

    result = vsc_to_raw(vsc_path)
    X, Y, Z = result['shape']
    print(f'OK    {vsc_path.name}  →  {result["raw"].name}  ({X}x{Y}x{Z})')
    converted += 1

print(f'\nDone — converted: {converted}, skipped: {skipped}')
```
